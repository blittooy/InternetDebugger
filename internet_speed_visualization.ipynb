```python
from mysql_db import MySQLDB
import datetime
import pandas as pd
import matplotlib.pyplot as plt
```

### Retreving Data From MySQL DB

```python
query = "Select * from internet_speed where download_speed < 1000"
df = pd.DataFrame(MySQLDB.pi_mysql_db().get_data(query), columns=['time', 'ping', 
                                                                 'download_speed', 
                                                                  'upload_speed'])
df = df.set_index('time')
```

### Internet Speed Profile

```python
fig, ax = plt.subplots(3, figsize=[14,8], sharex=True)
fig.subplots_adjust(hspace=0)
ax[0].plot(df['download_speed'], color='lightblue')
ax[0].set_title("Internet Speed Profile", fontsize=16)
ax[0].legend(['download_speed'], loc=1, fontsize=14)
ax[0].set_ylabel('mb/s', fontsize=14)
ax[1].plot(df['upload_speed'], color='indianred')
ax[1].legend(['upload_speed'], loc=1, fontsize=14)
ax[1].set_ylabel('mb/s', fontsize=14)
ax[2].plot(df['ping'], color='lightgreen')
ax[2].legend(['ping'], loc=1, fontsize=14)
t = ax[2].set_ylabel('ms', fontsize=14)
```
